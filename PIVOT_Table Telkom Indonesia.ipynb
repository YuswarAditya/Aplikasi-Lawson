# **Packages & Modules**

```python
import pandas as pd
import numpy as np
```

# **Load Data**

```python
#declares the result of previous step (cleansing)
df = pd.read_csv('/content/drive/MyDrive/Magang/DRI Batch 5/Report WO per 3 days/results/main/sep_jan2.csv')
df
```

# **Pivot WO Daily**

```python
wo_daily_jan = pd.pivot_table(df[df['MONTH']=='January'],index=['DATE'],
                              columns=['TERITORI'],aggfunc={'KONTAK':'count'})
wo_daily_jan
```

```python
wo_daily_des = pd.pivot_table(df[df['MONTH']=='December'],index=['DATE'],
                              columns=['TERITORI'],aggfunc={'KONTAK':'count'})
wo_daily_des
```

# **Pivot PS Daily**

```python
df['KESIMPULAN'].unique()
```

```python
ps_daily_jan = pd.pivot_table(df[(df['MONTH'] == 'January') & (df['KESIMPULAN']=='COMPLETED PS')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'KONTAK':'count'})
ps_daily_jan
```

```python
ps_daily_des = pd.pivot_table(df[(df['MONTH'] == 'December') & (df['KESIMPULAN']=='COMPLETED PS')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'KONTAK':'count'})
ps_daily_des
```

# **Pivot Kendala Daily**

```python
kendala_daily_jan = pd.pivot_table(df[(df['MONTH'] == 'January') & (df['KESIMPULAN']=='KENDALA')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'KONTAK':'count'})
kendala_daily_jan
```

```python
kendala_daily_des = pd.pivot_table(df[(df['MONTH'] == 'December') & (df['KESIMPULAN']=='KENDALA')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'KONTAK':'count'})
kendala_daily_des
```

# **Pivot Teknisi Daily**

```python
teknisi_daily_jan = pd.pivot_table(df[(df['MONTH'] == 'January')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'TEKNISI': pd.Series.nunique})
#the result of pivot has to multiplied by two
#because a team consists of 2 people
teknisi_daily_jan = teknisi_daily_jan*2
teknisi_daily_jan
```

```python
teknisi_daily_des = pd.pivot_table(df[(df['MONTH'] == 'December')],
                              index=['DATE'],columns=['TERITORI'],aggfunc={'TEKNISI': pd.Series.nunique})
#the result of pivot has to multiplied by two
#because a team consists of 2 people
teknisi_daily_des = teknisi_daily_des*2
teknisi_daily_des
```

# **PS/WO Daily**

```python
ps_wo_daily_jan = ps_daily_jan/wo_daily_jan
ps_wo_daily_jan.fillna(0,inplace=True)
ps_wo_daily_jan
```

```python
ps_wo_daily_des = (ps_daily_des/wo_daily_des)
ps_wo_daily_des.fillna(0,inplace=True)
ps_wo_daily_des
```

```python
#get the average value of each data
avg_ps_wo_jan = pd.DataFrame(ps_wo_daily_jan.mean()).reset_index()
avg_ps_wo_des = pd.DataFrame(ps_wo_daily_des.mean().reset_index())
#rename column
avg_ps_wo_jan.rename(columns = {0:'PS / WO'}, inplace = True)
avg_ps_wo_des.rename(columns = {0:'PS / WO'}, inplace = True)
```

```python
avg_ps_wo_jan
```

```python
avg_ps_wo_des
```

# **Kendala/WO Daily**

```python
kendala_wo_daily_jan = kendala_daily_jan/wo_daily_jan
kendala_wo_daily_jan.fillna(0,inplace=True)
kendala_wo_daily_jan
```

```python
kendala_wo_daily_des = kendala_daily_des/wo_daily_des
kendala_wo_daily_des.fillna(0,inplace=True)
kendala_wo_daily_des
```

```python
#get the average value of each data
avg_kendala_wo_jan = pd.DataFrame(kendala_wo_daily_jan.mean()).reset_index()
avg_kendala_wo_des = pd.DataFrame(kendala_wo_daily_des.mean().reset_index())
#rename column
avg_kendala_wo_jan.rename(columns = {0:'Kendala / WO'}, inplace = True)
avg_kendala_wo_des.rename(columns = {0:'Kendala / WO'}, inplace = True)
```

```python
avg_kendala_wo_jan
```

```python
avg_kendala_wo_des
```

```python
#merge with the previous average data
avg_jan = pd.merge(avg_ps_wo_jan, avg_kendala_wo_jan, how = 'outer')
avg_jan
```

```python
#merge with the previous average data
avg_des = pd.merge(avg_ps_wo_des, avg_kendala_wo_des, how = 'outer')
avg_des
```

# **PS/Teknisi Daily**

```python
ps_teknisi_daily_jan = pd.DataFrame(ps_daily_jan.values/teknisi_daily_jan.values)
ps_teknisi_daily_jan.insert(loc=0, column='DATE', value=wo_daily_jan.index)
ps_teknisi_daily_jan.columns = ['DATE', 'Banyumanik', 'Candi','Johar','Kendal',
                                'Majapahit','Salatiga','Simpang Lima','Tugu','Ungaran']
ps_teknisi_daily_jan.fillna(0,inplace=True)
ps_teknisi_daily_jan
```

```python
ps_teknisi_daily_des = pd.DataFrame(ps_daily_des.values/teknisi_daily_des.values)
ps_teknisi_daily_des.insert(loc=0, column='DATE', value=wo_daily_des.index)
ps_teknisi_daily_des.columns = ['DATE', 'Banyumanik', 'Candi','Johar','Kendal',
                                'Majapahit','Salatiga','Simpang Lima','Tugu','Ungaran']
ps_teknisi_daily_des.fillna(0,inplace=True)
ps_teknisi_daily_des
```

```python
#get the average value of each data
avg_ps_teknisi_jan = pd.DataFrame(ps_teknisi_daily_jan.mean()).reset_index()
avg_ps_teknisi_des = pd.DataFrame(ps_teknisi_daily_des.mean().reset_index())
#rename column
avg_ps_teknisi_jan.rename(columns = {0:'PS / Teknisi'}, inplace = True)
avg_ps_teknisi_des.rename(columns = {0:'PS / Teknisi'}, inplace = True)
```

```python
#merge with the previous average data
avg_jan['PS / Teknisi'] = avg_ps_teknisi_jan['PS / Teknisi']
#delete unneeded column
del avg_jan['level_0']
avg_jan
```

```python
#merge with the previous average data
avg_des['PS / Teknisi'] = avg_ps_teknisi_des['PS / Teknisi']
#delete unneeded column
del avg_des['level_0']
avg_des
```

# **Saving Result**

```python
ps_wo_daily_jan.to_csv('ps_wo_daily_jan.csv')
ps_wo_daily_des.to_csv('ps_wo_daily_des.csv')
kendala_wo_daily_jan.to_csv('kendala_wo_daily_jan.csv')
kendala_wo_daily_des.to_csv('kendala_wo_daily_des.csv')
ps_teknisi_daily_jan.to_csv('ps_teknisi_daily_jan.csv',index=False)
ps_teknisi_daily_des.to_csv('ps_teknisi_daily_des.csv',index=False)
avg_jan.to_csv('avg_jan.csv',index=False)
avg_des.to_csv('avg_des.csv',index=False)
```

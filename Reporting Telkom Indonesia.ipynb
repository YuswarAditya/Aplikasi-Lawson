# **Status Column (Make Kesimpulan & Main WO Column)**

```python
data['STATUS'].unique()
```

```python
#make kesimpulan column with the result of categorize value status column
conditions = [
    (data['STATUS'] == "CLOSE USAGE"),
    (data['STATUS'] != "CLOSE USAGE"),
    ]

values = ["COMPLETED PS","KENDALA"]

data['KESIMPULAN'] = np.select(conditions, values)
data[['STATUS','KESIMPULAN']]
```

```python
#make main wo column with the result of categorize value status column
conditions = [
    (data['STATUS'] == "ODP BLM GOLIVE") | (data['STATUS'] == "ODP FULL") | (data['STATUS'] == "ODP GENDONG") |
    (data['STATUS'] == "ODP LOSS/RETI") | (data['STATUS'] == "TANAM TIANG") | (data['STATUS'] == "TDK ADA ODP") |
    (data['STATUS'] == "UNDERGROUND"),

    (data['STATUS'] == "BATAL") | (data['STATUS'] == "BEDA ALAMAT PASANG") | (data['STATUS'] == "DOUBLE INPUT"),

    (data['STATUS'] == "ALAMAT TDK DITEMUKAN") | (data['STATUS'] == "PENDING") | (data['STATUS'] == "RESCHEDULE") |
    (data['STATUS'] == "RUKOS"),

    (data['STATUS'] == "CEK ALPRO") | (data['STATUS'] == "KENDALA TEKNISI"),

    (data['STATUS'] == "HR ONLY") | (data['STATUS'] == "HR+ONT-ON")
    ]

values = ["K.JARINGAN","K.ADM","K.PELANGGAN","K.TEKNISI","HR-KENDALA"]

data['MAIN WO'] = np.select(conditions, values)

for i in range(data.shape[0]):
  if data['MAIN WO'][i] == '0':
    data['MAIN WO'][i] = "COMPLETED PS"

data[['STATUS','KESIMPULAN','MAIN WO']]
```

# **Jam Input GD Column (Make Time Slot WO Column)**

```python
#make waktu input column with integer hour value of jam input column
data['WAKTU INPUT GD'] = pd.to_datetime(data['JAM INPUT GD (JJ.MM)'], format='%H:%M:%S').dt.hour
print('Missing Value : ', data['WAKTU INPUT GD'].isnull().sum())
```

```python
data[['JAM INPUT GD (JJ.MM)','WAKTU INPUT GD']]
```

```python
#categorize waktu input column with function
data["TIME SLOT WO"] = " "

def session(x):
  if (x >= 8) and (x < 11):
    return 'Pagi'
  elif (x >= 11) and (x < 14 ):
    return 'Siang'
  elif (x >= 14) and (x < 17):
    return 'Sore'
  elif (x >= 17) and (x < 22):
    return 'Malam'
  else :
    return 'Luar Jam Kerja'

for i in range(data.shape[0]):
  data["TIME SLOT WO"][i] = session(data["WAKTU INPUT GD"][i])

data[['JAM INPUT GD (JJ.MM)','WAKTU INPUT GD','TIME SLOT WO']]
```
